```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
#bokeh
import numpy as np

from bokeh.plotting import figure, output_file, show

# Total population, N.
N = 1000
# Initial number of infected and recovered individuals, I0 and R0.
I0, R0 = 1, 0
# Everyone else, S0, is susceptible to infection initially.
S0 = N - I0 - R0
# Contact rate, beta, and mean recovery rate, gamma, (in 1/days).
beta, gamma = 0.2, 1./10 
# A grid of time points (in days)
t = np.linspace(0, 160, 160)

# The SIR model differential equations.
def deriv(y, t, N, beta, gamma):
    S, I, R = y
    dSdt = -beta * S * I / N
    dIdt = beta * S * I / N - gamma * I
    dRdt = gamma * I
    return dSdt, dIdt, dRdt

# Initial conditions vector
y0 = S0, I0, R0
# Integrate the SIR equations over the time grid, t.
ret = odeint(deriv, y0, t, args=(N, beta, gamma))
S, I, R = ret.T

# Plot the data on three separate curves for S(t), I(t) and R(t)
fig = plt.figure(facecolor='w')
ax = fig.add_subplot(111, axisbelow=True)
ax.plot(t, S/1000, 'b', alpha=0.5, lw=2, label='Susceptibilidade')
ax.plot(t, I/1000, 'r', alpha=0.5, lw=2, label='Infectados')
ax.plot(t, R/1000, 'g', alpha=0.5, lw=2, label='Recuperados com imunidade')
ax.set_xlabel('Tempo /dias')
ax.set_ylabel('NÃºmero (1000s)')
ax.set_ylim(0,1.2)
ax.yaxis.set_tick_params(length=0)
ax.xaxis.set_tick_params(length=0)
ax.grid(b=True, which='major', c='w', lw=2, ls='-')
legend = ax.legend()
legend.get_frame().set_alpha(0.5)
for spine in ('top', 'right', 'bottom', 'left'):
    ax.spines[spine].set_visible(False)
    plt.savefig("SIR.png",dpi=360,format='png',quality=100)
# plt.show()

#bokeh plot
p = figure(title="SIR MODEL",
           x_range=(0, 5), y_range=(0, 1.2),
           background_fill_color="#fafafa")

p.line(t, S/1000, legend_label="Susceptibilidade",
       line_color="tomato", line_dash="dashed")
p.line(t, I/1000, legend_label="Infectados",
       line_color="gold", line_dash="dashed")
p.line(t, R/1000, legend_label="Recuperados com imunidade",
       line_color="olivedrab", line_dash="dashed")


p.legend.location = "top_left"

show(p)
```
